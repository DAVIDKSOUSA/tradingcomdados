### Bibliotecas a serem utilizadas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Carregando a base contendo os dados fundamentalistas

```python
base = pd.read_excel('IBOV_FUND_NEW.xlsx')
```

### Vamos visualizar a estrutura da base

```python
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)

base.head()
```

```python
base.dtypes
```

```python
sns.set(font_scale=2.5)
corr = base.corr()
plt.figure(figsize=(40,40))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right')
```

## Vamos mostrar um subset de um setor

```python
construtoras = base[base['Setor'] == "Construção e Engenharia"]
```

```python
construtoras.head()
```

```python
plt.figure(figsize=(40,40))
construtoras.sort_values('EBIT', 
                         ascending=False)[['Ticker','EBIT']].set_index('Ticker').plot.bar(stacked=True,figsize=(20,10))
```

```python
plt.figure(figsize=(40,40))
construtoras.sort_values('Margem Bruta', 
                         ascending=False)[['Ticker','Margem Bruta']].set_index('Ticker').plot.bar(stacked=True,
                                                                                                  figsize=(20,10))
```

```python
plt.figure(figsize=(40,40))
construtoras.sort_values('ROE', 
                         ascending=True)[['Ticker','ROE']].set_index('Ticker').plot.bar(stacked=True,figsize=(20,10))
```

```python
plt.figure(figsize=(40,40))
construtoras.sort_values('Dívida Líquida', 
                         ascending=True)[['Ticker','Dívida Líquida']].set_index('Ticker').plot.bar(stacked=True,
                                                                                                   figsize=(20,10))
```

```python
plt.figure(figsize=(40,40))
construtoras.sort_values('Receita Líquida', 
                         ascending=False)[['Ticker','Receita Líquida']].set_index('Ticker').plot.bar(stacked=True,
                                                                                                     figsize=(20,10))
```
